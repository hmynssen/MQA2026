## QuestĂŁo 1.1
Com $E = T + V = \frac{mv^2}{2} + mgy$, temos:

$$\frac{dE}{dt} = \frac{d}{dt}\left(\frac{m}{2} \vec{v} \cdot \vec{v}\right) + \frac{d}{dt}(mgy)$$

Aplicando a regra da cadeia:
$$\frac{dE}{dt} = \frac{m}{2} \cdot 2\vec{v} \cdot \frac{d\vec{v}}{dt} + mg\frac{dy}{dt}$$
$$\frac{dE}{dt} = m\vec{v} \cdot \vec{a} + mg\dot{y}$$

Como $\vec{a} = -g \hat{y}$:
$$\frac{dE}{dt} = m(\dot{x} \cdot 0 + \dot{y} \cdot (-g)) + mg\dot{y}$$
$$\frac{dE}{dt} = -mg\dot{y} + mg\dot{y} = 0$$

**ConclusĂŁo:** $\frac{dE}{dt} = 0$, logo a energia ĂŠ conservada.


## QuestĂŁo 1.2
Para um potencial $V(\vec{x})$ onde $\vec{F} = m\vec{a} = -\nabla V(\vec{x})$:

$$\frac{dE}{dt} = \frac{d}{dt} \left( \frac{1}{2}m\vec{v}^2 + V(\vec{x}) \right)$$
$$\frac{dE}{dt} = m\vec{v} \cdot \vec{a} + \nabla V(\vec{x}) \cdot \frac{d\vec{x}}{dt}$$

Substituindo $\nabla V(\vec{x}) = -m\vec{a}$ e $\frac{d\vec{x}}{dt} = \vec{v}$:
$$\frac{dE}{dt} = m\vec{v} \cdot \vec{a} + (-m\vec{a}) \cdot \vec{v} = 0$$

**ConclusĂŁo:** A energia se conserva para qualquer potencial conservativo.
