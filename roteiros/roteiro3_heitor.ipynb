## E&R Questão 3.13

O elétron é ambos e manifesta propriedades tanto de partícula quanto de onda, a depender do arranjo experimental utilizado para observá-lo. Este fenômeno é conhecido como Dualidade Onda-Partícula. Exemplos:

1. Comportamento ondulatório - quando o elétron se propaga pelo espaço sem ser perturbado por uma medição localizada, ele se comporta como uma onda.
2. Comportamento particular - no momento em que o elétron colide com o detector, a função de onda colapsa e a interação ocorre de forma perfeitamente localizada.


## E&R Problemas 3.8

Usando python para fazer um gráfico descente ao invés de Latex, segue no bloco abaixo o resultado.

Para o elétron temos que:

$$K = \frac{p^2}{2m_0}$$
$$p = \sqrt{2m_0K}$$

E assim a equação de comprimento de onda fica:

$$\lambda_e = \frac{h}{\sqrt{2m_0K}}$$

Com $m_0$ sendo a massa de repouso do elétron.
Agora, para o fóton, de forma muito mais simples, já que $K = E = pc$, temos que:

$$\lambda_f = \frac{hc}{K}$$



```python
import numpy as np
import matplotlib.pyplot as plt


h = 4.135667696e-15 ## eV*s
c = 299792458       ## m/s
m0_c2 = 0.511e6     ## Energia de repouso do elétron em eV
m0 = m0_c2 / (c**2) ## Massa do elétron em eV/(m/s)^2
hc_ev_nm = 1240.0   ## eV * nm -> para tirar o lambda em nm

K_min = 0.001 * m0_c2
K_max = 0.05 * m0_c2
K = np.linspace(K_min, K_max, 500)

lambda_eletron = hc_ev_nm / np.sqrt(2 * m0_c2 * K)
lambda_foton = hc_ev_nm / K

plt.figure(figsize=(9, 6), dpi=100)
plt.style.use('seaborn-v0_8-whitegrid')

plt.plot(K / m0_c2 * 100, lambda_eletron, 
    label='Elétron', color='#1f77b4', 
    linewidth=2.5)
plt.plot(K / m0_c2 * 100, lambda_foton, 
    label='Fóton', color="#ca640b", 
    linestyle='--', linewidth=2.5)

plt.xlabel('$K$ (em % de $m_0c^2$ do elétron)', fontsize=11)
plt.ylabel('$\\lambda$ (nm)', fontsize=11)

plt.legend(fontsize=11, frameon=True)
plt.tight_layout()
plt.show()
```
