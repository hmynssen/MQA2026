## E&R Problema 1.16
A radiância espectral é dada por:
$$R_T = \int_0^{\infty} \frac{2\pi h}{c^2} \frac{\nu^3}{e^{h\nu/kT} - 1} d\nu$$

Fazendo a mudança de variáveis $x = \frac{h\nu}{kT} \implies d\nu = \frac{kT}{h}dx$:
$$R_T = \frac{2\pi h}{c^2} \left(\frac{kT}{h}\right)^4 \int_0^{\infty} \frac{x^3}{e^x - 1} dx$$

Substituindo o valor da integral tabelada:
$$R_T = \frac{2\pi h}{c^2} \left(\frac{kT}{h}\right)^4 \left(\frac{\pi^4}{15}\right) = \left( \frac{2\pi^5 k^4}{15 c^2 h^3} \right) T^4 = \sigma T^4$$


## O calor específico dos gases

Temos os seguintes graus de liberdade:
- Translação $\implies E_{\text{trans}} = \frac{3}{2}kT$
- Rotação $\implies E_{\text{rot}} = 2 \cdot \frac{1}{2}kT = kT$
- Vibração $\implies E_{\text{vib}} = kT$

Isso resulta em uma energia total $U = \frac{7}{2}RT$, implicando um calor específico constante $c_v = \frac{\partial U}{\partial T} = \frac{7}{2}R$, independente da temperatura. 

Entretanto, falta levar em consideração que:
1. a radiação eletromagnética é quantizada;
2. a absorção e excitação dos graus de liberdade de rotação e vibração são quantizadas;
3. cada um dos 3 tipos de graus de liberdade possui energias fundamentais mínimas distintas.

Levando em conta que a probabilidade de excitação térmica de um estado quântico com energia $\Delta E$ é governada pelo fator de Boltzmann ($e^{-\Delta E/kT}$), analisamos os estados em 3 casos possíveis:

1. Baixas temperaturas:
   a energia térmica média é muito menor que o primeiro espaçamento energético quântico de rotação e de vibração ($kT \ll \Delta E_{\text{rot}} \ll \Delta E_{\text{vib}}$). A probabilidade de excitação desses estados é desprezível ($e^{-\Delta E/kT} \to 0$). Apenas a translação absorve energia:
   $$U = \frac{3}{2}RT \implies \hat{c}_v = \frac{3}{2}R$$

2. Médias temperaturas:
   a energia térmica supera o gap rotacional ($kT > \Delta E_{\text{rot}}$), ativando a rotação. porém, a vibração permanece desprezível:
   $$U = \frac{3}{2}RT + RT = \frac{5}{2}RT \implies \hat{c}_v = \frac{5}{2}R$$

3. Altas temperaturas:
   a energia do sistema é alta o suficiente para transpor os níveis quantizados de vibração ($kT > \Delta E_{\text{vib}}$), fazendo com que o gás atinja o limite clássico completo da equipartição:
   $$U = \frac{7}{2}RT \implies \hat{c}_v = \frac{7}{2}R$$


## A radiação de corpo negro em 1D

A partir das condições de contorno de uma cavidade unidimensional de comprimento $a$ (ver), os modos normais vibracionais da radiação eletromagnética possuem frequências quantizadas por $\nu = \frac{nc}{2a}$. O número de modos no intervalo $d\nu$, contabilizando os dois estados de polarização da luz, é:
$$N(\nu)d\nu = 2 \left(\frac{2a}{c}\right)d\nu = \frac{4a}{c}d\nu$$

A densidade de modos por unidade de comprimento é dada por $n_{1D}(\nu)d\nu = \frac{4}{c}d\nu$.

Assumindo o teorema da equipartição da energia ($\bar{\mathcal{E}} = kT$):
$$\rho_{1D}(\nu)d\nu = \frac{4}{c}kT d\nu$$

Que é a expressão de Rayleigh-Jeans para uma dimensão.
