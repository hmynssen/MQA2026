## Questão 1.1
Com $E = T + V = \frac{mv^2}{2} + mgy$, temos:

$$\frac{dE}{dt} = \frac{d}{dt}\left(\frac{m}{2} \vec{v} \cdot \vec{v}\right) + \frac{d}{dt}(mgy)$$

Aplicando a regra da cadeia:
$$\frac{dE}{dt} = \frac{m}{2} \cdot 2\vec{v} \cdot \frac{d\vec{v}}{dt} + mg\frac{dy}{dt}$$
$$\frac{dE}{dt} = m\vec{v} \cdot \vec{a} + mg\dot{y}$$

Como $\vec{a} = -g \hat{y}$:
$$\frac{dE}{dt} = m(\dot{x} \cdot 0 + \dot{y} \cdot (-g)) + mg\dot{y}$$
$$\frac{dE}{dt} = -mg\dot{y} + mg\dot{y} = 0$$

**Conclusão:** $\frac{dE}{dt} = 0$, logo a energia é conservada.


## Questão 1.2
Para um potencial $V(\vec{x})$ onde $\vec{F} = m\vec{a} = -\nabla V(\vec{x})$:

$$\frac{dE}{dt} = \frac{d}{dt} \left( \frac{1}{2}m\vec{v}^2 + V(\vec{x}) \right)$$
$$\frac{dE}{dt} = m\vec{v} \cdot \vec{a} + \nabla V(\vec{x}) \cdot \frac{d\vec{x}}{dt}$$

Substituindo $\nabla V(\vec{x}) = -m\vec{a}$ e $\frac{d\vec{x}}{dt} = \vec{v}$:
$$\frac{dE}{dt} = m\vec{v} \cdot \vec{a} + (-m\vec{a}) \cdot \vec{v} = 0$$

**Conclusão:** A energia se conserva para qualquer potencial conservativo.

## Questão 1.6
Considerando $x(t) = A \cos(\omega t + \phi)$ e $v(t) = -A\omega \sin(\omega t + \phi)$, com $\omega^2 = \frac{k}{m}$:

$$E = \frac{1}{2} m v(t)^2 + \frac{1}{2} k x(t)^2$$
$$E = \frac{1}{2} m A^2 \omega^2 \sin^2(\omega t + \phi) + \frac{1}{2} k A^2 \cos^2(\omega t + \phi)$$

Substituindo $m\omega^2 = k$:
$$E = \frac{1}{2} k A^2 [\sin^2(\omega t + \phi) + \cos^2(\omega t + \phi)]$$
$$E = \frac{kA^2}{2}$$

## Questão 1.8

Primeiro, calculamos as derivadas em relação ao tempo:


$$\frac{\partial^2 s(u)}{\partial t^2} = \frac{\partial}{\partial t} \left( \frac{d s(u)}{du} \cdot \frac{\partial u}{\partial t} \right)$$

Por regra do produto:
$$\frac{\partial^2 s(u)}{\partial t^2} =\frac{d s(u)}{du}\cdot\frac{\partial^2 u}{\partial t^2} + \frac{d}{du}\left(\frac{\partial s(u)}{\partial t}\right)\cdot \frac{\partial u}{\partial t}$$

Como $\frac{\partial u}{\partial t} = $ cte, e $\frac{\partial^2 u}{\partial t^2} = 0$:
$$\frac{\partial^2 s(u)}{\partial t^2} =\frac{d}{du}\left(\frac{\partial s(u)}{\partial t}\right)\cdot \frac{\partial u}{\partial t}$$

$$=\frac{d}{du}\left(\frac{d s(u)}{du}\cdot \frac{\partial u}{\partial t}\right)\cdot \frac{\partial u}{\partial t}$$

$$=\left(\frac{d^2 s(u)}{du^2}\cdot \frac{\partial u}{\partial t} + \frac{d s(u)}{du}\cdot \frac{\partial }{\partial t}\frac{\partial u}{\partial u}\right)\cdot \frac{\partial u}{\partial t}$$

$$=\left(\frac{d^2 s(u)}{du^2}\cdot \frac{\partial u}{\partial t} + 0\right)\cdot \frac{\partial u}{\partial t}$$

E por fim:
$$\frac{\partial^2 s(u)}{\partial t^2}=\frac{d^2 s(u)}{du^2}\cdot \left(\frac{\partial u}{\partial t}\right)^2$$


Analogamente para o espaço:
$$\frac{\partial^2 s(u)}{\partial x^2} = \frac{d^2 s(u)}{du^2} \cdot \left(\frac{\partial u}{\partial x}\right)^2$$

Substituindo na equação de onda:
$$s''(u) \cdot \omega^2 = c^2 \cdot s''(u) \cdot k^2$$
Como $\omega = ck$, temos $\omega^2 = c^2 k^2$, o que resulta em $1 = 1$ .

## Questão 1.9
Usando $\sin(a\pm b)  = \cos(a)\cos(b) \pm \sin(a)\sin(b)$, e tomando $b=\omega t + \delta_t$ e $a=kx+\delta_x$, temos que:

$$\frac{1}{2} \sin(a+b)-\frac{1}{2} \sin(a-b) = \sin(a)\sin(b)$$

$$\frac{1}{2} \sin(kx + \delta_x + \omega t + \delta_t) - \frac{1}{2} \sin(kx + \delta_x -\omega t + \delta_t  ) = \sin(kx+\delta_x) \cdot \sin(\omega t + \delta_t)$$

Onde do lado esquerdo temos as ondas propagantes.

## Questão 1.11
Segue diretamente que:

$$u + iv = e^{i\varphi}z = a \cos(\phi) +ia\sin(\phi)+ib\cos(\phi)-b\sin(\phi)$$

De onde tiramos $u=a \cos(\phi)-b\sin(\phi)$ e $v=ia\sin(\phi)+ib\cos(\phi)$

Em forma matricial:
$$\begin{pmatrix} u \\ v \end{pmatrix} = \begin{pmatrix} \cos \varphi & -\sin \varphi \\ \sin \varphi & \cos \varphi \end{pmatrix} \begin{pmatrix} a \\ b \end{pmatrix}$$

Esta é a matriz de Rotação 2D, que rotaciona um vetor no plano complexo por um ângulo $\varphi$.

## Questão 1.14
Para um sistema de $N$ massas sem paredes, a força sobre a massa $i$ devido aos vizinhos é:
$$F_i = k(x_{i+1} - x_i) - k(x_i - x_{i-1}) = k(x_{i+1} - 2x_i + x_{i-1})$$

Isso pode ser escrito como $\ddot{\vec{x}}= -\frac{k}{m} \mathbf{A} \vec{x}$, onde a matriz $\mathbf{A}$ possui componentes:
$$A_{ij} = \begin{cases} 2, & \text{se } i=j \\ -1, & \text{se } i=j \pm 1 \\ 0, & \text{caso contrário} \end{cases}$$

Como $A_{ij} = A_{ji}$, a matriz é auto-adjunta. Note que para as bordas $A_{00}=A_{NN}=1$, porém isso não altera o resultado.
